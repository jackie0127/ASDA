```python
def easy_task_3_analysis(df, df_name):
    """Prints a simple summary of missing values and duplicates."""
    print(f"\nMissing values and duplicates: {df_name}")

    total_rows = len(df)
    total_cols = df.shape[1]

    # 1. Missing Values Analysis
    missing_counts = df.isnull().sum()
    total_missing_values = missing_counts.sum()

    # Filter to show only columns with missing data
    missing_data_summary = pd.DataFrame({
        'Missing count': missing_counts,
        '% Missing': (missing_counts / total_rows * 100).round(2)
    })
    missing_data_summary = missing_data_summary[missing_data_summary['Missing count'] > 0]

    print("\nMissing values per column:")
    if missing_data_summary.empty:
        print("No missing values found.")
    else:
        print(missing_data_summary.to_markdown())

    # 2. Overall Missing Summary
    percentage_dataset_affected = (total_missing_values / (total_rows * total_cols)) * 100

    print(f"\nTotal missing values: {total_missing_values}")
    print(f"Percentage of dataset affected: {percentage_dataset_affected:.2f}%")

    # 3. Duplicates Analysis
    duplicated_rows_found = df.duplicated().sum()
    percentage_rows_affected = (duplicated_rows_found / total_rows) * 100

    print(f"Duplicated rows found: {duplicated_rows_found}")
    print(f"Percentage of rows in dataset affected: {percentage_rows_affected:.2f}%")
    print("--------------------------------------------------")
```

```python
easy_task_3_analysis(eq_df, "GEOFON Earthquake Data (eq_df)")
easy_task_3_analysis(weather_df, "Seattle Weather Data (weather_df)")
```
